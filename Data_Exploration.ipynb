```python
import pandas as pd
import numpy as np

data_by_year = {}
relevant_cols = ["cik", "ticker", "filing_date", "item1a_risk", "item7_mda"]

key = "20"
for i in range(9,21):
    filestub = "0" + str(i) if i <10 else str(i)
    filestub = key + filestub
    data_by_year[filestub] = pd.read_csv(filestub + ".csv", usecols=relevant_cols)
    data_by_year[filestub] = data_by_year[filestub].dropna(subset=['cik', 'item1a_risk', 'item7_mda']).drop_duplicates()
```
